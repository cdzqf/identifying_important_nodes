## Kendall's Tau of  synthetic networks

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-ticks')
plt.rcParams['xtick.direction'] = 'in'
plt.rcParams['ytick.direction'] = 'in'
plt.rc('text', usetex=True)
plt.rc('font', family='Times New Roman')
font_size = 14
plt.rc('xtick', labelsize=font_size)
plt.rc('ytick', labelsize=font_size)
```

### Since KS is not applicable to BA scale-free networks, it is not considered here

```python
filename = ["ER", "BA"]          
name_labels = ["(a) ER", "(b) BA"]
ymin = [0.7, 0.5]
ymax = [0.9, 0.9]
plt.figure(figsize=(6, 2.5))
for i, name in enumerate(filename):
    results = np.loadtxt("./Kendall_tau_results/Kendall_tau_"+name+".dat")
    plt.subplot(1, 2, i+1)
    plt.title(name_labels[i] + " " + name[3:])
    plt.plot(results[:,0], results[:,1], "ro-", markersize=3, label="DC")
    # plt.plot(results[:,0], results[:,2], "bs-", markersize=3, label="KS")
    plt.plot(results[:,0], results[:,3], "g^-", markersize=3, label="CC")
    plt.plot(results[:,0], results[:,4], "yv-", markersize=3, label="EC")
    plt.plot(results[:,0], results[:,5], "mH-", markersize=3, label="DCGM")
    plt.plot(results[:,0], results[:,6], "c*-", markersize=3, label="GGC")
    plt.plot(results[:,0], results[:,7], color="pink", marker="P", markersize=3, label="EHCC")
    plt.plot(results[:,0], results[:,8], "ko-", markersize=3, label="DC$^{+}$")
    plt.plot(results[:,0], results[:,9], "ks-", markersize=3, label="DCGM$^{+}$")
    plt.plot(results[:,0], results[:,10], "kD-", markersize=3, label="DCGM$^{++}$")
    plt.xlim([0.4, 1.6])
    plt.ylim([ymin[i],ymax[i]])
    plt.xlabel(r"$\beta/\beta_c$", fontsize=font_size)
    if i==0:
        plt.ylabel(r"$\tau$", fontsize=font_size)
        plt.legend(loc=8, ncol=3, columnspacing=1.0, fontsize=8)
        
plt.tight_layout(pad=0.5, w_pad=1.3, h_pad=1.5)
# plt.savefig("Kendall_tau_R2_synthetic.pdf")
```
